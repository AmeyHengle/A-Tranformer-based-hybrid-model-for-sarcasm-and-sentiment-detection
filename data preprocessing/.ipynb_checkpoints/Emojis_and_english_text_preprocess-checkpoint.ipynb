```python
filew = open("../dataset/emoticons_mapping.txt","w",encoding="utf8")
file3 = open("../dataset/emojis.txt","r",encoding="utf8")
file4 = open("../dataset/arabic_emoji.txt","r",encoding="utf8")
file5 = open("../dataset/emojitext.txt","r", encoding="utf8")

emoticon_df = pd.DataFrame(columns = ["emoticon", "english_translation","arabic_translation"])
emojis = []
english = []
arabic = []
for emoji, en, ar in zip(file3.readlines(), file5.readlines(), file4.readlines()):
    emojis.append(emoji.replace("\n",""))
    english.append(en.replace("\n",""))
    arabic.append(ar.replace("\n",""))
    filew.write(emoji.replace("\n","") + " " + en.replace("\n","") + " " + ar.replace("\n","") + "\n")
file3.close()
file4.close()
file5.close()
```

```python
def get_emoji_count(text, emoji_set):
    emoji_cnt = 0
    for x in text.split():
        if x in UNICODE_EMO or x in EMOTICONS:
            emoji_cnt += 1
    return emoji_cnt

def remove_english(text):
    output = re.sub(r'\s*[A-Za-z0-9]+\b', '' , text)
    return output.rstrip()

def find_emojis(text):
    return re.findall('[\U00010000-\U0010ffff]',text)

# Function for converting emojis into word
def convert_emojis(text):
    for emot in UNICODE_EMO:
        text = text.replace(emot, "_".join(UNICODE_EMO[emot].replace(",","").replace(":","").split()))
    return text
```
