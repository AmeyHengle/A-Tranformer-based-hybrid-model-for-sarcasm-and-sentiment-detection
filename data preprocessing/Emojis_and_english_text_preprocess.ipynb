```
import numpy as np
import pandas as pd
import re
```

```
data = pd.read_csv('/content/preprocessed_1.csv')
data['tweet'][:50]
```

```
def find_emojis(text):
    return re.findall('[\U00010000-\U0010ffff]',text)
```

```
list_of_emojis = list(filter(None, data['tweet'].apply(find_emojis)))
list_of_emojis
```

```
def remove_english(text):
    output = re.sub(r'\s*[A-Za-z0-9]+\b', '' , text)
    return output.rstrip()
```

```
data['tweet'] = data['tweet'].apply(remove_english)
```

```
data.to_csv('/content/preprocessed_2.csv')
```
