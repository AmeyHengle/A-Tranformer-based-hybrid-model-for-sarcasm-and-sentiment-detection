```
import numpy as np
import pandas as pd
import re
```

```
data = pd.read_csv('/content/preprocessed_fin.csv')
data.head()
```

```
sentiment_values = pd.read_csv('/content/arabic-sentiment-lexicons.csv')
sentiment_values.head()
```

```
trial = data['tweet'][0].split()
for word in trial:
    print(word in sentiment_values['Arabic Term'])
```

```
for tweet in data['tweet']:
    if type(tweet) == str:
        pos_count, neg_count = 0,0
        sentiment_score = 0.0
        for word in tweet.split():
            if type(word) == str:
                if word in sentiment_values['Arabic Term']:
                    print(word)
                    pos_count += sentiment_values.query(f'Arabic Term=={word}')['Positive Count']
                    neg_count += sentiment_values.query(f'Arabic Term=={word}')['Negative Count']
                    sentiment_score += sentiment_values.query(f'Arabic Term=={word}')['Sentiment Score']
        data['total_pos_count'] = pos_count
        data['total_neg_count'] = neg_count
        data['total_sentiment_score'] = sentiment_score
```
