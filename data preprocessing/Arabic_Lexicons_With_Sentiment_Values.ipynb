```
import numpy as np
import pandas as pd
import re
```

```
data = pd.read_csv('/content/preprocessed_fin.csv')
data.head()
```

```
sentiment_values = pd.read_csv('/content/arabic-sentiment-lexicons.csv')
sentiment_values.head()
```

```
tweets = data['tweet'].tolist()
terms = sentiment_values['Arabic Term'].tolist() 
negative_counts = sentiment_values['Negative Count'].tolist()
positive_counts = sentiment_values['Positive Count'].tolist()
sentiment_scores = sentiment_values['Sentiment Score'].tolist()
```

```
pos_count_values = []
neg_count_values = []
sen_values = []

for tweet in tweets:
    if type(tweet) == str:
        pos_count, neg_count = 0,0
        sen_score = 0.0
        i = 0
        for word in tweet.split():
            if word in terms:
                i = terms.index(word)
                pos_count += positive_counts[i]
                neg_count += negative_counts[i]
                sen_score += sentiment_scores[i]
        pos_count_values.append(pos_count)
        neg_count_values.append(neg_count)
        sen_values.append(sen_score)
    else:
        pos_count_values.append(0)
        neg_count_values.append(0)
        sen_values.append(0)


data['total_pos_count'] = pos_count_values
data['total_neg_count'] = neg_count_values
data['total_sentiment_score'] = sen_values
```
